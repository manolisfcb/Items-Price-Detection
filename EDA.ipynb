```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib
import json
matplotlib.rcParams['figure.figsize'] = (20, 10)
```

```python
def build_dataset():
    data = [json.loads(x) for x in open("MLA_100k_checked_v3.jsonlines")]
    target = lambda x: x.get("condition")
    N = -10000
    X_train = data[:N]
    X_test = data[N:]
    y_train = [target(x) for x in X_train]
    y_test = [target(x) for x in X_test]
    for x in X_test:
        del x["condition"]
    return X_train, y_train, X_test, y_test
```

```python
data_train, y_train, data_test, y_test = build_dataset()
```

```python
def replace_reputacion(x):
    x = str.lower(x)
    # remove accents
    x = x.replace('á', 'a')
    x = x.replace('é', 'e')
    x = x.replace('í', 'i')
    x = x.replace('ó', 'o')
    x = x.replace('ú', 'u')
    
    if 'reputacion' in x:
        return 'reputacion'
    else:
        return x
    
df_train['warranty'] = df_train['warranty'].apply(lambda x: replace_reputacion(x))
```

```python
def waranty_to_numeric(x):
    if x == 'unknown' or x == 'reputacion' :
        return 0
    return 1

df_train['warranty'] = df_train['warranty'].apply(lambda x: waranty_to_numeric(x))
```
